```python
def countNegatives(grid):
    rows = len(grid)
    cols = len(grid[0])
    total_negatives = 0

    for row in grid:
        left, right = 0, cols - 1
        first_negative_index = cols   # assume no negative initially

        while left <= right:
            mid = (left + right) // 2

            if row[mid] < 0:
                first_negative_index = mid
                right = mid - 1      # search left for first negative
            else:
                left = mid + 1

        # all elements after first_negative_index are negative
        total_negatives += cols - first_negative_index

    return total_negatives


# ðŸ”¹ Driver code
grid = [
    [4, 3, 2, -1],
    [3, 2, 1, -1],
    [1, 1, -1, -2],
    [-1, -1, -2, -3]
]

print("Total negative numbers:", countNegatives(grid))
```
